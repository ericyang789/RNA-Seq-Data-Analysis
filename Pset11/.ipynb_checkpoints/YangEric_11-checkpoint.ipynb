## 1. reproduce Watsons's K-means result

Watson claims that there are 8 distinct cell types. Let's run K-means and try and group the cells based on the RNA-Seq data.

```python
# Adapted from Anonymous Student One's K-means script: "w05-answers-student1"
# I made minor changes to the 'update' and 'kmeans' functions to achieve a lower sum of the squared distance
def euclid(a, b):
    """
    FUNCTION to calculate euclidean distance given two vector arrays, a & b
    Can also calculate in logspace based on global variable EUCLID
    Inputs: a, b = Arrays
    Output: distance = Distance between vectors
    """
    distance = 0.
    for i in range(len(a)):
        distance += (a[i] - b[i]) ** 2
    return np.sqrt(distance)

def assign (X, centroids):
    """
    FUNCTION that assigns X points to centroids 
    Inputs: X = Array of points, centroids = array of centroid coordinates
    Outputs: clusters = Array of index of clusters, tot_sqdist as Float
    """
    clusters = np.zeros(len(X)).astype(int) # Create array for groupings
    tot_sqdist = 0 # Initialize total square distance calculation that we will keep track of
    
    # Enumerate over all points to assign them and calculate tot_sqdist
    for i, point in enumerate(X):
        opt_distance = euclid(point, centroids[0]) # Initialize optimum distance for single point
        opt_centroid = 0 # Initialize optimum centroid index
        
        # Calculate closest centroid to point
        for centroid in range (1, K):
            distance = euclid(point, centroids[centroid])
            if distance < opt_distance:
                opt_distance = distance
                opt_centroid = centroid
                
        # Set point to optimum centroid and add to tot_sqdist measure
        clusters[i] = opt_centroid
        dist = opt_distance 
        tot_sqdist += dist ** 2
    
    return clusters, tot_sqdist

def update (X, K, clusters):
    """
    FUNCTION that updates centroids given clustering clusters and points X
    Inputs: X = Array of points, clusters = index of cluster
    Outputs: New centroid coordinates, number of observations in each cluster
    """
    n,p = np.shape(X)
    
    # Assign arrays to store Ck and new centroid coordinates
    cluster_count = np.zeros(K)
    new_centroids = np.zeros((K, p))
    
    # Summing across numerators and denominators for calulations
    for i in range(n):
        index = clusters[i]
        cluster_count[index] += 1 # Denominator
        for j in range(p):
            new_centroids[index, j] += X[i][j] # Numerator
    
    # Average position of each point assigned to each cluster to get the updated centroids
    for i in range(K):
        if cluster_count[i] > 0: # If cluster has no points, ignore 
            new_centroids[i] = np.divide(new_centroids[i], cluster_count[i])

    return new_centroids, cluster_count

def kmeans (X, K):
    """
    FUNCTION that implements hard k-means algorithim with euclidean distance
    Inputs: X = points to cluster, K = number of clusters
    Output: centroids = coordinates of centroids, clusters = index map of points to cluster
            fraction = percentage of observations in each cluster, tot_sqdist = euclidean metric to minimize
    """
    # Inner function that determines when to stop iteration
    def stop (a, b):
        count = 0
        for i in range(len(a)):
            if a[i] != b[i]:
                count += 1
        return (count == 0)
    
    idx = np.random.choice(len(X), K)
    centroids = X[idx]
    
    # Initiate k-means algorithim once
    clusters, tot_sqdist = assign(X, centroids)
    centroids, cluster_count = update(X, K, clusters)
    old_clusters = clusters
    
    # Continue iteration until there is no difference    
    while True:
        clusters, tot_sqdist = assign(X, centroids)
        if stop(old_clusters, clusters):
            break
        centroids, cluster_count = update(X, K, clusters)
        old_clusters = clusters
    
    # Calculate the "fraction" in Wiggin's table
    fraction = np.divide(cluster_count, np.sum(cluster_count))
    
    return centroids, clusters, fraction, tot_sqdist
```

```python
# Extract data parameters
X = data.values
n,p = np.shape(X) # n is number of cells, p is number of genes
log_X = np.log(X)
```

We will run K-means using the log of the RNA-seq counts. Here we accept the K-means' equal variance and uniform mass assumptions as Watson did, but we are free to pre-process the data as we see fit without affecting the integrity of K-means. To be consistent with Watson, we represent the data as log counts. Note that the K-means algorithm can sometimes get stuck in local optima depending on the initialization, so here we will try 10 different initializations by assigning random points as starting centroids. For each initialization, the we define convergence as no change in point assignment. We then pick the initialization that leads to the smallest sum of squared distance. The 10 random initializations below take awhile to run!

```python
K = 8 # number of clusters
EUCLID = True
best_seed = -1
best_dist = float("inf")
for run in range(35,45):
    np.random.seed(run)
    centroids, clusters, fraction, tot_sqdist = kmeans(log_X, K)
    if (tot_sqdist < best_dist) and (len(np.unique(clusters)) == K): # we don't want orphan clusters
        best_seed = run
        best_dist = tot_sqdist
print("Best RNG seed {0:2d}:  totdist {1:12.1f}".format(best_seed, best_dist))
```

We achieve a very similar result, only slightly off from Watson's sum of squared distance of 87378.2. Again, this is expected as different initializations can vary results. Let's explore these assignments further

## 2. reduce the dimensionality

Before running PCA with SVD, we first center the data so that the first PC does not come out to be proportional to the mean of the data over rows.

```python
# center data
means = np.mean(log_X, axis=0)
centered = log_X-means

# SVD
U,S,W_T = np.linalg.svd(centered, full_matrices=True) # W_T here ordered by descending eigenvalue

# project data 
Y = U @ np.diag(S)
```

Let's make sure the decomposed matrices make sense mathematically with the following properties: 
1. The total variance of the data calculated by SVD should be the same as the variance obtained via eigendecomposition.
2. All eigenvectors obtained via SVD should be orthogonal to one another. 
3. Each eigenvector obtained via SVD should be unit vectors. 

```python
# check total variance can be achieved by analyzing S matrix
sigma = np.cov(log_X, rowvar = False)
lamb,W = np.linalg.eigh(sigma) # W here ordered by ascending eigenvalue
assert np.round(np.sum(lamb)) == np.round(np.sum(S**2/(n-1))) 

# check eigenvectors are orthogonal, dot product of any eigenvectors should be 0
for j in range(p):
    for k in range(j+1,p):
        assert np.round(np.dot(W_T.T[j],W_T.T[k]),3) == 0

# check eigen vectors are unit vectors
for j in range(p):
    assert np.round(np.linalg.norm(W_T.T[j]),3) == 1
```

We passed all the tests, it looks like our SVD worked. Let's look at how the cells project onto the first 2 PCs.

```python
# plot all 200 cells in 2D expression space using their projections onto the first two principal axes
plt.figure(figsize=(8,8))
plt.scatter(Y[:, 0], Y[:, 1])
plt.xlabel('projection onto component 1')
plt.ylabel('projection onto component 2')
plt.axis('square')
plt.show()
```

By looking at the projected data onto the first 2 PCs, Watson was right to expect 8 clusters. There seems to be 8 distinct groups of cells that hold certain combinations of PC1-PC2 pairs. By lowering the dimensions and only visualizing for the components that drive the most variation, we better understand how the cells and clusters differ. When reducing dimensions, there is a nuance in determining which components hold valuable information regarding variation, and which components can be considered as noise. Let's see if it was fair of us to only consider 2 components by looking at the eigenvalues for each component and see how that compares to eigenvalues from a simulated negative control dataset.

It seems like K-means is missing the mark in clustering the cells in PC1-PC2 space. This makes sense as in part 1, we used all of the dimensions for clustering. In part 2, we found that PCs 3-200 are essentially noise but collectively they account for more than 90% of the data's variance. Therefore, K-means here incorporated a lot of irrelevant information that don't drive the data's variance into clustering the cells. To properly assign cells in PC space, we can tell k-means to only consider the first two components of our projected data, which are most inpactful in driving cell clusters. Again we run multiple initializations here to try and find the global optimum.

```python
# use first two PCs of projected data to cluster
K = 8 # number of clusters
best_seed = -1
best_dist = float("inf")
for run in range(100):
    np.random.seed(run)
    centroids, clusters, fraction, tot_sqdist = kmeans(Y[:,0:2], K)
    if tot_sqdist < best_dist:
        best_seed = run
        best_dist = tot_sqdist
print("Best RNG seed {0:2d}:  totdist {1:12.1f}".format(best_seed, best_dist))
```

With this approach, K-means was able to identify the 8 clusters in PC space. To further test my hypothesis, let's see how K-means perform when we use the first 10 components.

## 4. reconstruct the expression patterns

Let's reconstruct our original data with only the first 2 prinicpal components as to "denoise" the data.

```python
# reconstruct data with first 2 PCs
recon = U[:,0:2] @ np.diag(S[0:2]) @ W_T[0:2,:]
```

```python
# rearrange cells to order by clusters
ordered = [] # for gene vs cell heatmap
breaks = []
counts = 0
for clust in range(K):
    for idx, assign in enumerate(clusters):
        if assign == clust:
            ordered.append(recon[idx])
            counts += 1 
    breaks.append(counts)
ordered = np.array(ordered)
```

```python
# plot heat map
plt.figure(figsize=(40,20))
plt.imshow(ordered, cmap='Purples')
plt.hlines(breaks,-0.5,1999.5,colors='orange')
plt.colorbar(fraction=0.005)
plt.ylim(0,200)
plt.xlabel('gene #')
plt.ylabel('cell #')
```

```python
# plot only first 10 genes
# plot heat map
plt.figure(figsize=(50,10))
plt.imshow(ordered[:,0:11], aspect="auto", cmap='Purples')
plt.hlines(breaks,-0.5,10.5,colors='orange')
plt.colorbar()
plt.ylim(0,200)
plt.xlabel('gene #')
plt.ylabel('cell #')
```

It is still a bit difficult to visualize the clusters via heatmaps. The first challenge is that there are so many dimensions, making it difficult to study the counts for each gene and cluster while also maintaining a wholistic view of the heatmap. The second challenge is that when there are a few several outlier values, the colorbar has to extend its range to accomodate for them. This pushes the bulk of the data all into one color bin when in reality there are distinguishable differences among them. When we only plot 10 of the 2001 genes, we not only zoom in but we also readjust the colorbar to a smaller range, making it possible to distinguish the clusters via a heatmap. The first problem is difficult to fix given that our device only has so many pixels so it's difficult to visualize a large dimensioned heatmap with our naked eyes. Let's see if we can fix the second challenge if we normalize the data by the range each gene so that the outlier blocks in the heatmap don't drag out the colorbar range.

```python
ordered_norm = ordered/np.ptp(ordered,axis=0)
plt.figure(figsize=(40,20))
plt.imshow(ordered_norm, cmap='Purples')
plt.hlines(breaks,0,2000,colors='orange')
plt.colorbar(fraction=0.005)
plt.ylim(0,200)
plt.xlabel('gene #')
plt.ylabel('cell #')
```

```python
# plot only first 10 genes
# plot heat map
plt.figure(figsize=(50,10))
plt.imshow(ordered_norm[:,0:11], aspect="auto", cmap='Purples')
plt.hlines(breaks,-0.5,10.5,colors='orange')
plt.colorbar()
plt.ylim(0,200)
plt.xlabel('gene #')
plt.ylabel('cell #')
```

The clusters look much more distinguishable even without zooming into a subset of genes. Indeed, normalizing the data by each gene help tremendously for heatmap visualization.
Let's take a look at how the influential genes drive the clusters.

```python
# plot influential genes
plt.figure(figsize=(50,10))
plt.imshow(ordered_norm[:,influential],aspect="auto", cmap='Purples')
plt.hlines(breaks,-0.5,len(influential)-0.5,colors='orange')
plt.colorbar(fraction=0.5)
plt.ylim(0,200)
plt.xlabel('gene #')
plt.ylabel('cell #')
```

As expected, there is a clear difference in gene expression among clusters for each of the influential genes. These were the genes contributed the most to the variation in PC1 and PC2. The takeaway for this pset is that PCA is extremely useful to visualize high dimensional data and tease out what is driving the variation behind our data.
