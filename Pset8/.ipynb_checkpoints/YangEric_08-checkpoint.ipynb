```python
import pandas            as pd
import scipy.optimize    as optimize
import scipy.stats       as stats
import numpy             as np
import matplotlib.pyplot as plt
import math
```

```python
# read in data
data = pd.read_csv('w08-data.tbl', delim_whitespace=True, index_col=0)
```

```python
# get sigmas
pm_sigmas = data.loc['gene']
sigmas = [int(s[2:]) for s in pm_sigmas] # remove '+-' to get just the numbers

# get times
tmp = [x.replace('hr.1','') for x in data.columns] # remove units to get just the numbers
times = [int(x.replace('hr','')) for x in tmp] 

# remove formatting and sigma lines
data = data.iloc[2:]
data.loc[:,'4hr':'24hr.1'] = data.loc[:,'4hr':'24hr.1'].astype(float) # convert string to float 

data.head()
```

## 1. solve by maximum likelihood

```python
# define negative log-likelihood based on harmonic regression model
def nll(params, t, y, sigmas):
    '''
    params is a np array length 3 containing a, b, phi
    t is a np array of length 8 containing all the time points of each experiment
    y is a np array of length 9 containing tpm values
    sigmas is a np array of length 8 containing std devs of each experiment
    '''
    a, b, phi = params 
    a = np.exp(a) # makes sure a is positive, need to exponentiate out later
    mu = b + a * np.sin(2.*math.pi*(1/24)*(t+phi))
    ll_terms = stats.norm.logpdf(y - mu, loc=0, scale=sigmas)
    return -ll_terms.sum()
```

```python
# initialize guesses for a, b, phi
init = np.array([1,1,1]) 

# optimize for each gene
ngenes = data.shape[0]
a = np.empty(ngenes)    # a that gives best nll for each gene
b = np.empty(ngenes)    # b that gives best nll for each gene
phi = np.empty(ngenes)  # phi that gives best nll for each gene
nlls = np.empty(ngenes) # optimized nlls of each gene
for i in range(data.shape[0]):
    min_results = optimize.minimize(nll, init, (times, data.iloc[i], sigmas))
    a[i] = np.exp(min_results.x[0])
    b[i] = min_results.x[1]
    phi[i] = min_results.x[2]%24
    nlls[i] = min_results.fun
```

The opimized negative log likelihoods (nll) for each gene is stored in the 'nll' array. 'a','b', and 'phi' arrays stores the parameters that give the optimal nll for each gene.

```python
print('a:', a)
print('b:', b)
print('phi', phi)
```

Our results are in the same order of magnitude compared to Moriarty's. All of our amplitude values are positive and all of our phase values are between 0 and 24 hours, which make biological sense!

## 2. compare solutions

```python
# extract parameters from moriarty results
moriarty = ! python moriarty.py 'w08-data.tbl'
all_params = np.empty((12,3))# 12 lists of a,b,phi, one for each gene
for i in range(ngenes):
    a_m = float(moriarty[i+2][21:26]) 
    b_m = float(moriarty[i+2][14:19])
    phi_m = float(moriarty[i+2][28:33])
    all_params[i] = np.array([np.log(a_m),b_m,phi_m]) # log 'a' here since 'a' gets exp'd in nll function

# calculate moriarty total log likelihood
moriarty_ll = 0
for i in range(ngenes):
    moriarty_ll += nll(all_params[i], times, data.iloc[i], sigmas)
    
# calculate likelihood ratio (ours divided by Moriarty's)
likelihood_ratio = np.exp(-nlls.sum()-(-moriarty_ll))
likelihood_ratio
```

After comparing the total log likelihood summed over all 12 genes, it looks like my solution is 3.68 x $10^{103}$ more likely than Moriarty's. It is common for likelihood ratios to be very large given the exponentiation. I will explore the difference between the two models in part 3. 

## 3. plot the fits

```python
# set up subplots
f = plt.figure(figsize=(12,8))
for i in range(ngenes):
    ax = f.add_subplot(3,4,i+1)
    
    # raw data
    ax.errorbar(times,data.iloc[i],sigmas,fmt="s",label='raw data')
    
    x = np.linspace(0,25,100)
    # our model
    y = b[i] + a[i] * np.sin(2.*math.pi*(1/24)*(x+phi[i]))
    ax.plot(x,y,label='M-L')
    
    # Moriarty's model
    y_m = all_params[i,1] + np.exp(all_params[i,0]) * np.sin(2.*math.pi*(1/24)*(x+all_params[i,2]))
    ax.plot(x,y_m,label='Moriarty')
    ax.set_title(data.index[i])
    ax.set_xlabel('time (hr)')
    ax.set_ylabel('expression (tpm)')
handles, labels = ax.get_legend_handles_labels()
f.legend(handles, labels, loc='upper left')
f.tight_layout(pad=1.0)
```

In his harmonic regression method, Moriarty assumes a normal noise distribution with an uniform standard deviation for each gene-timepoint measurement. This is a mistake, as biological experiments have different reliabilities due to different replicate sizes and technical noise. It is necessary to account for the different variability of each gene-timepoint pair, which is known in this case. Because Moriarty did not consider this, he treats that highly variable measurements as reliable as the ones with lower variance, thereby giving them equal contribution to his harmonic regression model. This is most noticeable in the 'kiwi', 'tangerine', 'huckleberry', 'lentil' and 'cauliflower' genes. Looking at the plots for these genes, Moriarty's model gets pulled towards the "outlier" datapoints with high variability while our model is more resistant to those points. Our model accounts for the variance differences and restricts the contributions of highly variant points. This is why knowing the variances, our model is more likely than Moriarty's, who assumes uniform variability. The takeaway here is that if we are given known variability information in biological experiments, like we are here, it is often a good idea to take that into consideration when generating models from the data.
