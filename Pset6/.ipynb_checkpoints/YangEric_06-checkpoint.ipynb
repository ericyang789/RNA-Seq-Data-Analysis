```python
import numpy as np
import scipy.stats as stats
import scipy.special as special
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. the beginner's game

For the beginner's game, we are trying to figure out the mean of a normal distribution, given a sample from the normal distribution and the true standard deviation of the distribution. The possible means are given as discrete integers.

```python
# all possible mus and sigmas from dartboard
board_mus = list(range(-100,101,10))
board_sigmas = list(range(5,101,5))
```

```python
def beginner(X, known_sigma, possible_mus):
    '''
    Inputs:
    - X is a list of observations drawn from some normal distribution 
    - known_sigma is an integer that is the std dev of that normal dist.
    - possible_mus is a list of integers, with one of those values being the mean of the normal dist
    
    calculates posterior probability for each possible mu, p(mu|X, sigma) in regular and logspace
    '''
    
    # calculate post prob for each possible mu
    probs = np.zeros(len(possible_mus))
    log_probs = np.zeros(len(possible_mus))
    for idx, mu in enumerate(possible_mus):
        numerator = 1 
        log_numerator = 0 # also calculate probs in logspace
        for x in X:
            numerator *= stats.norm.pdf(x,loc=mu,scale=known_sigma) 
            log_numerator += stats.norm.logpdf(x,loc=mu,scale=known_sigma) 
        probs[idx] = numerator * (1/21) 
        log_probs[idx] = log_numerator + np.log(1/21) 
        
    # normalize, divide by marginal (subtract in logspace)
    denominator = sum(probs)
    log_denominator = special.logsumexp(log_probs)
    probs = [p / denominator for p in probs]
    log_probs = [p - log_denominator for p in log_probs]
    
    return probs, log_probs
```

```python
# pub's calculated prob distribution, adapted from student-game.py
def bar_beginner(X, sigma, mu_values):
    """ 
    Given an ndarray X_1..X_n, and a known sigma;
    and a list of the mu values in each column;
    return a list of the inferred P(mu | X,sigma) for each column.
    """
    xbar = np.mean(X)
    N    = len(X)
    Pr   = [ stats.norm.pdf(x, loc=xbar, scale= sigma / np.sqrt(N)) for x in mu_values ]  # proportional to std error of the mean
    Z    = sum(Pr)                   # normalization constant
    Pr   = [ p / Z for p in Pr ]     # normalization to a discrete probability distribution
    return Pr 
```

```python
# Bayes' post probs n=3
beginner_probs_3, beginner_log_probs_3 = beginner([11.50, -2.32, 9.18],60,board_mus)
```

```python
# Bar probs, n=3
bar_beginner_probs_3 = bar_beginner([11.50, -2.32, 9.18],60,board_mus)
```

```python
# plots post probs on semilog scale
plt.semilogy(board_mus,beginner_probs_3)
plt.semilogy(board_mus,bar_beginner_probs_3)
plt.title('posterior prob of mu given data and sigma, n=3')
plt.xlabel('mu')
plt.ylabel('posterior probability')
plt.legend(["posterior probs","bar_calculations"])
```

## 2. the advanced game

For the advanced game, we are trying to figure out the mean and standard deviation of a normal distribution, given a sample from the normal distribution. The possible means and standard deviations are given as discrete integers.

```python
def advanced(X, possible_sigmas, possible_mus):
    '''
    Inputs:
    - X is a list of observations drawn from some normal distribution 
    - possible_sigmas is a list of integers, with one of those values being the sd of the normal dist
    - possible_mus is a list of integers, with one of those values being the mean of the normal dist
    
    calculates posterior probability for each possible mu and sigma combination, p(mu,sigma|X)
    '''
    # calculate post prob for each possible mu and sigma combination
    probs = np.zeros((len(possible_sigmas),len(possible_mus)))
    log_probs = np.zeros((len(possible_sigmas),len(possible_mus)))
    for sig_idx, sigma in enumerate(possible_sigmas):
        for mu_idx, mu in enumerate(possible_mus):
            numerator = 1 
            log_numerator = 0 # also calculate probs in logspace
            for x in X:
                numerator *= stats.norm.pdf(x,loc=mu,scale=sigma) 
                log_numerator += stats.norm.logpdf(x,loc=mu,scale=sigma) 
            probs[sig_idx,mu_idx] = numerator * (1/21) 
            log_probs[sig_idx,mu_idx] = log_numerator + np.log(1/21) 
        
    # normalize, divide by marginal (subtract in logspace)
    denominator = np.sum(probs)
    log_denominator = special.logsumexp(log_probs)
    for sig_idx, prob in enumerate(probs):
        for mu_idx, p in enumerate(prob):
            probs[sig_idx,mu_idx] /= denominator
    for sig_idx, prob in enumerate(log_probs):
        for mu_idx, p in enumerate(prob):
            log_probs[sig_idx,mu_idx] -= log_denominator
    
    return probs, log_probs
```

```python
# adapted from student-game.py
def bar_advanced(X, mu_values):
    """ 
    Given an ndarray X_1..X_n,
    and a list of the mu values in each column;
    return a list of the inferred P(mu | X) for each column.
    """
    xbar = np.mean(X)
    s    = np.std(X, ddof=1)     # note that numpy.sd() by default calculates a population std dev; to get sample std. dev., set ddof=1
    N    = len(X)
    Pr   = [ stats.norm.pdf(x, loc=xbar, scale= s / np.sqrt(N)) for x in mu_values ]  # proportional to std error of the mean
    Z    = sum(Pr)                   # normalization constant
    Pr   = [ p / Z for p in Pr ]     # normalization to a discrete probability distribution
    return Pr
```

```python
# what the bar should be using to calculate odds: t-distribution
def probdist_t(X, mu_values):
    """ 
    Given an ndarray X_1..X_n,
    and a list of the mu values in each column;
    return a list of the inferred P(mu | X) for each column,
    according to Student's t distribution with N-1 degrees of freedom.
    """
    N    = len(X)
    xbar = np.mean(X)
    s    = np.std(X, ddof=1)
    t    = [ (xbar - mu) / (s / np.sqrt(N)) for mu in mu_values ]    # t statistic, given sample mean, sample stddev, and N
#   t    = [ stats.ttest_1samp(X, mu)[0] for mu in mu_values ]       # ... (equivalently, python can calculate t statistic for you)
    Pr   = [ stats.t.pdf(val, N-1) for val in t ]
    Z    = sum(Pr)
    Pr   = [ p / Z for p in Pr ]    
    return Pr
```

## 3. where's the advantage?

For the beginner's game, when the population standard deviation is given, the two methods are essentially identical, regardless of sample size. As a result, each method gives the sample mean the highest probability for being the true mean and uses the given population standard deviation to assign probabilities for the remaining possible true mean values. Mathematically, the bar calculations computes pseudo-Z-scores for every potential true mean $\mu_k$, where $ Z_{pseudo} = \frac{\mu_k - \bar{X}}{\sigma / \sqrt{n}}$. Since we know the population standard deviation, the denominator becomes a constant, and the only source of variation comes from the numerator term dictated by the sample mean, which results in a normal Z distribution with mean as the sample mean and sd as $ {\sigma / \sqrt{n}} $ after normalization. Our Bayes calculation calculates Z-scores for every potential true mean where $ Z = \sum_{i=1}^n \frac{\bar{X}_i - \mu_k}{\sigma / \sqrt{n}} $ Knowing the population standard deviation, the denominator is again constant, resulting in a normal Z distribution with mean as sample mean and sd as $ {\sigma / \sqrt{n}} $ after normalization. These Z-distributions are identical which explains the identical posterior probabilities. The bar calculates its odd correctly for the beginner's game, and there is no advantage for us here.

For the advanced game,
